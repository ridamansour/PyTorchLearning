```python
pure_time_data =[]

for idx, dur in enumerate(durations):
    if dur.find("|")!=-1:
        pure_time_data.append("0min")
        print(pure_time_data[-1])
    else:
        pure_time_data.append(dur)
```

```python
import re
from datetime import timedelta

def parse_duration(text):
    hours = minutes = 0
    match = re.search(r'(\d+)\s*hr', text)
    if match:
        hours = int(match.group(1))
    match = re.search(r'(\d+)\s*min', text)
    if match:
        minutes = int(match.group(1))
    return timedelta(hours=hours, minutes=minutes)
```

```python
df = pd.DataFrame({
   "section_num": sections,
    "Vid idx": list(map(lambda x: -1 if x.find(". ")==-1 else int(x[:x.find(".")]), titles)),
    "title": list(map(lambda x: x[x.find(". ")+2:] if x.find(". ")!=-1 else x, titles)),
    "duration" : list(map(lambda x: parse_duration(x), pure_time_data)),
    "done" : [False]*len(titles)
})
```
